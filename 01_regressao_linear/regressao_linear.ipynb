```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
import scipy.stats as statsmodel
import seaborn as sns
```

```python
base = pd.read_csv('mt_cars.csv')
base.shape
```

```python
base.head()
```

```python
corr =base.corr()
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt='.2f')
```

```python
column_pairs = [('mpg','cyl'),('mpg','disp'), ('mpg','hp'),('mpg','wt'),('mpg','drat'),('mpg','vs')]
n_plots = len(column_pairs)
gig, axes = plt.subplots(nrows=n_plots, ncols=1, figsize=(6,4 * n_plots))

for i, pair in enumerate(column_pairs):
  x_col, y_col = pair
  sns.scatterplot(x=x_col, y=y_col,data=base, ax=axes[i])
  axes[i].set_title(f'{x_col} vs {y_col}')

plt.tight_layout()
plt.show()
```
